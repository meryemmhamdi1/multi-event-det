````python
from flexnlp.integrations.flexnlp import PlainTextIngester
from vistautils.misc_utils import WithId

pipeline_builder = Pipeline.builder()
pipeline_builder.add(PlainTextIngester())
doc = Document.from_text('MY_SAMPLE_DOC_ID', 'hello world')
pipeline_builder.add(AssignLanguageAnnotator('eng'))
pipeline_builder.add(SpacyAnnotator.create_for_language(
                  SpacyAnnotator.ENGLISH, use_regions=False,
                  respect_existing_sentences=False))
pipeline = pipeline_builder.build()

doc_text_and_id = WithId('MY_SAMPLE_DOC_ID', 'hello world')
output = pipeline.process(doc_text_and_id)
````
